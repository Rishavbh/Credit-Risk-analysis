### Missing data

```python
# created function to avoid repetition
def get_missing_data(df):
    missing_app_data = (df.isnull().sum()/len(df)) * 100
    missing_app_data = missing_app_data.rename_axis(['columns']).to_frame('% missing').reset_index().sort_values(by='% missing', ascending=False)
    missing_app_data['data_type'] = missing_app_data['columns'].apply(lambda col: str(app_data_sel[col].dtype))
    return missing_app_data
```

```python
#Creating a function for plotting pie charts
def show_pie_plot(df, cat_col, title='Pie Plot', size=(8,8)):
    # Pie chart, where the slices will be ordered and plotted counter-clockwise:
    df_val_cnt = df[cat_col].value_counts()
    labels = df_val_cnt.index
    sizes = df_val_cnt.values
#     explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
    fig1, ax1 = plt.subplots(figsize=size)
    fig1.suptitle(title)
    ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```

```python
def dist_plot(title, df_0, df_1, col, size = (15, 5)):
    fig, (ax1, ax2) = plt.subplots(1,2,figsize=size)
    fig.suptitle(title)
    ax1.title.set_text('No Default')
    ax2.title.set_text('Default')
    sns.distplot(df_0[col], ax=ax1, bins=30)
    sns.distplot(df_1[col], ax=ax2, bins=30)
    plt.show()
```

```python
def box_plot(title, df_0, df_1, col, size = (15, 5)):
    fig, (ax1, ax2) = plt.subplots(1,2,figsize=size)
    fig.suptitle(title)
    ax1.title.set_text('No Default')
    ax2.title.set_text('Default')
    sns.boxplot(y=df_0[col], ax=ax1)
    sns.boxplot(y=df_1[col], ax=ax2)
    plt.show()
```

#### Relation between numeric and categorical columns

```python
def combined_bar_plot(title, df, col_x, col_y, col_z, size = (15, 5), hue=None):
    new_col_z = col_z + '_Mean'
    df_grp = df.groupby([col_x, col_y])[col_z].mean().to_frame(new_col_z).reset_index()
    col_x_vals = df_grp[col_x].unique()
    num_subplots = len(col_x_vals)
    fig, axes = plt.subplots(num_subplots, figsize=size)
    fig.suptitle(title)
    for i in range(num_subplots):
        title = col_x_vals[i]
        ax_title = axes[i]
        ax_title.title.set_text(title)
        df_title = df_grp[df_grp[col_x] == title]
        sns.barplot(y=new_col_z, x=col_y, data=df_title, hue=hue, ax=ax_title, estimator=np.mean)
    plt.tight_layout()
    plt.show()
```

```python
def combined_count_plot(title, merged_df, col, size=(12,5), tight_layout=False):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=size)
    fig.suptitle(title)
    sns.countplot(merged_df[col], ax=ax1)
    t1 = merged_df[merged_df['TARGET'] == 1]
    t1 = t1[col].value_counts(normalize=True).rename_axis(col).to_frame('% Distribution with Target=1').reset_index()
    sns.barplot(x=col, y='% Distribution with Target=1', data=t1, ax=ax2)
    ax1.title.set_text('{} distribution on complete data'.format(col))
    ax2.title.set_text('{} % Distribution with Target=1'.format(col))
    if tight_layout:
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")
        plt.tight_layout()
    plt.show()
```
